```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('telecom_churn.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['State', 'Voice mail plan'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['International plan'] = data['International plan'].map({'Yes': 1, 'No': 0})
```

```python
data.head()
```

```python
data.info()
```

```python
y = data['Churn'].astype('int')
```

```python
y.dtype
```

```python
x = data.drop(['Churn'], axis=1)
```

```python
x.head()
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.3, random_state=17)
```

```python
x_train.shape, x_valid.shape
```

```python
tree_model = DecisionTreeClassifier(random_state=17)
```

```python
cross_val_score(tree_model, x_train, y_train, cv=5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_params = {'max_depth': range(1, 11), 'max_features': [0.5, 0.7, 1]}
```

```python
tree_grid = GridSearchCV(tree_model, tree_params, cv=5, n_jobs=-1)
```

```python
tree_grid.fit(x_train, y_train)
```

```python
tree_grid.best_score_
```

```python
tree_grid.best_params_
```

```python
tree_valid_res = tree_grid.predict(x_valid)
```

```python
tree_valid_res
```

```python
y_valid
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, tree_valid_res)
```
